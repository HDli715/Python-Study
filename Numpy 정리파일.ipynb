#### #2 예제

##### list 연산 데이터 처리)

```python
li1 = list(range(10000))
li2 = list(range(10000))

def add_list():    
    li3 = []
    for i in range(10000):
        li3.append(li1[i] + li2[i])
    return li3

add_list()
```

##### array 연산 데이터 처리)

```python
array1 = np.array(range(10000))
array2 = np.array(range(10000))

def add_array():
    array3 = array1 + array2
    return array3

add_array()
```

#### #3 속도 비교

```python
import timeit
```

```python
print(timeit.timeit(add_list , number = 10000))    # list는 느림
print(timeit.timeit(add_array , number = 10000))   # array가 더 빠름
```
